```python
def rules():
    print("RULES")
    print("-------------------------------------------------------------")
    print("you start with 5 lives")
    print("if you win you get an extra life ")
    print("if you lose you lose 1 life")
    print("if you draw you life count remains unchanged")
    print("-------------------------------------------------------------")
    print("MAKE SURE YOU DON'T USE CAPITALS")
    print("-------------------------------------------------------------")
    print("To see the rules type rules ")
    print("to exit the game at any point type exit")
    print("-------------------------------------------------------------")
```

```python
import random
lives = 5
won = "You won"
lose = "The computer wins"
draw = "you drew"
myList = ["rock","paper","scissor"]
compLife = 7
print("Welcome to ROCK PAPER SCISSORS")
print("-------------------------------------------------------------")
print("-------------------------------------------------------------")
rules()
print("The computer has lives as well ")
print("Can you beat the computer ??")
print("Good Luck")
print("-------------------------------------------------------------")
while(True):
    print("rock paper scissor ?")
    user = input("YOU:  ")
    if user == "rules":
        rules()
        continue
    elif user == "exit":
        break
    cc = random.choice(myList)
    print("COMP: ",cc)
    #----ROCK----
    if user == "rock":
        if cc == "rock":
            print(draw)
           
        elif cc == "scissor":
            print(won)
            
            lives += 1
            compLife -= 1
            print("YOUR LIVES:", lives)
            print("Comp LIVES:", compLife)
            if lives == 0:
                print("Computer won the game")
                print("---------------------------------------------")
                print("THANKYOU FOR PLAYING")
                break
                
            elif compLife == 0:
                print("Hurray! You Won")
                print("---------------------------------------------")
                print("THANKYOU FOR PLAYING")
                break
        else:
            print(lose)
            
            lives -= 1
            compLife += 1
            print("YOUR LIVES:", lives)
            print("Comp LIVES:", compLife)
            if lives == 0:
                print("Computer won the game")  
                print("---------------------------------------------")
                print("THANKYOU FOR PLAYING")
                break
            elif compLife == 0:
                print("Hurray! You Won")
                print("---------------------------------------------")
                print("THANKYOU FOR PLAYING")
                break
    #----PAPER----
    if user == "paper":
        if cc == "paper":
            print(draw)
           
        elif cc == "rock":
            print(won)
            
            lives += 1
            compLife -= 1
            print("YOUR LIVES:", lives)
            print("Comp LIVES:", compLife)
            if lives == 0:
                print("Computer won the game")   
                print("---------------------------------------------")
                print("THANKYOU FOR PLAYING")
                break
            elif compLife == 0:
                print("Hurray! You Won")
                print("---------------------------------------------")
                print("THANKYOU FOR PLAYING")
                break
        else:
            print(lose)
           
            lives -= 1
            compLife += 1
            print("YOUR LIVES:", lives)
            print("Comp LIVES:", compLife)
            if lives == 0:
                print("Computer won the game")
                print("---------------------------------------------")
                print("THANKYOU FOR PLAYING")
                break
            elif compLife == 0:
                print("Hurray! You Won")
                print("---------------------------------------------")
                print("THANKYOU FOR PLAYING")
                break
    #----SCISSOR---
    if user == "scissor":
        if cc == "scissor":
            print(draw)
           
        elif cc == "paper":
            print(won)
            
            lives += 1
            compLife -= 1
            print("YOUR LIVES:", lives)
            print("Comp LIVES:", compLife)
            if lives == 0:
                print("Computer won the game")
                print("---------------------------------------------")
                print("THANKYOU FOR PLAYING")
                break
            elif compLife == 0:
                print("Hurray! You Won")
                print("---------------------------------------------")
                print("THANKYOU FOR PLAYING")
                break
        else:
            print(lose)
            
            lives -= 1
            compLife += 1
            print("YOUR LIVES:", lives)
            print("Comp LIVES:", compLife)
            if lives == 0:
                print("Computer won the game")
                print("---------------------------------------------")
                print("THANKYOU FOR PLAYING")
                break
            elif compLife == 0:
                print("Hurray! You Won")
                print("---------------------------------------------")
                print("THANKYOU FOR PLAYING")
                break
```
